```python
def clean_txt(example):
    for txt in ["dialogue", "summary"]:
       example[txt]  = example[txt].lower()
       example[txt]  = example[txt].replace("\\", "")
       example[txt]  = example[txt].replace("/", "")
       example[txt]  = example[txt].replace("\n", "")
       example[txt]  = example[txt].replace("``", "")
       example[txt]  = example[txt].replace('"', '')
       example[txt]  = example[txt].replace("--", "")

    return example
```

```python
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens = True)

    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens = True)

    result = rouge.compute(predictions = decoded_preds, references = decoded_labels, use_stemmer = True)

    prediction_length = [np.count_nonzero(pred != tokenizer.pad_token_id) for pred in predictions]
    result["generated_length"] = np.mean(prediction_length)

    return {k: round(v, 4) for k, v in result.items()}
```

```python
!pip install pytorch
!pip install transformers
!pip install Seq2SeqTrainingArguments
from transformers import Seq2SeqTrainingArguments

!pip install evaluate
import evaluate
from transformers import Seq2SeqTrainingArguments

# Import torch here
import torch

def preprocess_function(examples):
  inputs = [prefix + doc for doc in examples["dialogue"]]
  model_inputs = tokenizer(inputs, max_length=1024, truncation=True)
  # Setup the tokenizer for targets
  with tokenizer.as_target_tokenizer():
    labels = tokenizer(examples["summary"], max_length=128, truncation=True)

  model_inputs["labels"] = labels["input_ids"]
  return model_inputs

tokenized_highlight_summary_ds = highlight_summary_ds.map(preprocess_function, batched=True)
tokenized_highlight_summary_ds = tokenized_highlight_summary_ds.remove_columns(highlight_summary_ds["train"].column_names)

# Define train_dataset and eval_dataset here
train_dataset = tokenized_highlight_summary_ds["train"]
eval_dataset = tokenized_highlight_summary_ds["test"]

# Define training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="highlight_summary_model_trained_on_reduced_data", # Moved output_dir here
    evaluation_strategy="epoch", # Moved evaluation_strategy here
    learning_rate=2e-5,  # Moved learning_rate here
    per_device_train_batch_size=2,  # Moved per_device_train_batch_size here
    per_device_eval_batch_size=2,  # Moved per_device_eval_batch_size here
    weight_decay=0.01,  # Moved weight_decay here
    save_total_limit=2,  # Moved save_total_limit here
    num_train_epochs=2,  # Moved num_train_epochs here
    predict_with_generate=True,  # Moved predict_with_generate here
    fp16=True,  # Moved fp16 here
    push_to_hub=True,  # Moved push_to_hub here
)
torch.cuda.empty_cache()

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics = compute_metrics
)
trainer.train()
```
